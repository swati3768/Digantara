# Assignment - Data Analyst (Conjunctions in space)

```python
#importing the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the data from the downloaded CSV file
df = pd.read_csv("C:/Users/swati/Downloads/sort-minRange (1).csv")
df
```

```python
df['TCA']=pd.to_datetime(df['TCA'])
df.set_index('TCA',inplace=True)
df.info()
```

```python
#A) Derive high level analytics from the whole data set for a single day. In other words, derive general analytics of the whole set of conjunction scenarios
```

```python
df["2023-11-10"]
```

```python
# Combine and organize the data for analysis
df = df.sort_values('TCA')  # Sort data by TCA in chronological order
df
```

```python
# Define the desired date for analysis
selected_day = '2023-11-10'  # Replace with the date you want to analyze

# Filter the data to select the rows for the specified date
selected_data = df[df.index.date == pd.to_datetime(selected_day).date()]

# Calculate the total number of conjunctions among active satellites for the selected day.
total_conjunctions = selected_data.shape[0]
total_conjunctions
```

```python
# Calculate descriptive statistics for TCA_RANGE, TCA_RELATIVE_SPEED, and MAX_PROB.
tca_range_stats = selected_data['TCA_RANGE'].describe()
tca_relative_speed_stats = selected_data['TCA_RELATIVE_SPEED'].describe()
max_prob_stats = selected_data['MAX_PROB'].describe()
print(tca_range_stats)
print(tca_relative_speed_stats)
print(max_prob_stats)
```

```python
# Count the number of conjunctions that exceed a specific dilution threshold (DILUTION).
dilution_threshold = 0.01  # Replace with your desired threshold
high_dilution_conjunctions = selected_data[selected_data['DILUTION'] > dilution_threshold]
num_high_dilution_conjunctions = high_dilution_conjunctions.shape[0]
num_high_dilution_conjunctions
```

```python
# Step 3: Visualization
plt.figure(figsize=(10, 6))

# Histograms for TCA_RANGE, TCA_RELATIVE_SPEED, and MAX_PROB
plt.subplot(2, 2, 2)
plt.hist(selected_data['TCA_RANGE'], bins=10, alpha=0.5, color='b', label='TCA_RANGE')
plt.title("TCA_RANGE Distribution")

plt.subplot(2, 2, 3)
plt.hist(selected_data['TCA_RELATIVE_SPEED'], bins=10, alpha=0.5, color='g', label='TCA_RELATIVE_SPEED')
plt.title("TCA_RELATIVE_SPEED Distribution")

plt.subplot(2, 2, 4)
plt.hist(selected_data['MAX_PROB'], bins=10, alpha=0.5, color='r', label='MAX_PROB')
plt.title("MAX_PROB Distribution")

plt.tight_layout()

plt.show()
```

```python
# Step 3: Visualization
plt.figure(figsize=(10, 6))

# Pie chart to show the proportion of conjunctions exceeding the dilution threshold
plt.figure(figsize=(8, 6))
labels = ['Below Threshold', 'Above Threshold']
sizes = [total_conjunctions - num_high_dilution_conjunctions, num_high_dilution_conjunctions]
colors = ['#ff9999', '#66b3ff']
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%')
plt.title("Conjunctions by Dilution Threshold")
plt.show()
```

```python
#B) Represent the conjunctions data of a single satellite or a satellite constellation.The analytics should be intuitive, represented in an easily understandable format andshould enable decision making from a satellite operatorâ€™s point of view.
```

```python
# Define the NORAD Catalog ID of the satellite or constellation of interest
specific_satellite_id = 49529 # Replace with the NORAD Catalog ID you want to analyze

# Filter the dataset for the specific satellite or constellation
satellite_data = df[(df['NORAD_CAT_ID_1'] == specific_satellite_id) | (df['NORAD_CAT_ID_2'] == specific_satellite_id)]

# Group the data by date to count the number of conjunctions on each day
conjunctions_by_date = satellite_data.groupby('TCA').size()
conjunctions_by_date
```

```python
# Create a timeline of conjunctions for the specific satellite or constellation
plt.figure(figsize=(12, 6))
plt.plot(conjunctions_by_date.index, conjunctions_by_date.values, marker='o', linestyle='-', color='b')
plt.title(f'Conjunctions for NORAD ID {specific_satellite_id}')
plt.xlabel('Date (TCA)')
plt.ylabel('Number of Conjunctions')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Create a scatter plot to visualize maximum probability and relative speed for conjunctions
plt.figure(figsize=(8, 6))
plt.scatter(satellite_data['MAX_PROB'], satellite_data['TCA_RELATIVE_SPEED'], alpha=0.5)
plt.title(f'Conjunctions for NORAD ID {specific_satellite_id}')
plt.xlabel('Maximum Probability')
plt.ylabel('TCA Relative Speed')
plt.grid(True)
plt.show()
```

```python
# Define the start and end dates for the 5-day period
start_date = pd.Timestamp('2023-11-06')
end_date = start_date + pd.DateOffset(days=4)

# Filter the DataFrame to include data only for the 5-day period
filtered_df = df[(df.index >= start_date) & (df.index <= end_date)]

# Create a line plot with 'TCA_RANGE' and 'TCA_RELATIVE_SPEED' columns for the 5-day period
plt.figure(figsize=(12, 6))
plt.plot(filtered_df.index, filtered_df['TCA_RANGE'], linestyle='-', marker='o', color='b', alpha=0.5, label='TCA_RANGE')
plt.plot(filtered_df.index, filtered_df['TCA_RELATIVE_SPEED'], linestyle='-', marker='o', color='g', alpha=0.5, label='TCA_RELATIVE_SPEED')
plt.title('TCA_RANGE and TCA_RELATIVE_SPEED Over the 5-Day Period')
plt.xlabel('Date')
plt.ylabel('Values')
plt.legend()  # Add a legend to differentiate the two lines
plt.grid(True)

# Rotate x-axis labels for readability (optional)
plt.xticks(rotation=45)

# Show the plot
plt.show()
```
